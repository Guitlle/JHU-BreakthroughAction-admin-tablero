```python
client = Client(cluster) 
```

```python
fullpersonas = dd.read_csv('/home/guillermo/DATOS/GTVitales y Censo/CENSO GT 2018.csv/PERSONA_BDP.csv', dtype=str)
```

```python
hogares = dd.read_csv('/home/guillermo/DATOS/GTVitales y Censo/CENSO GT 2018.csv/HOGAR_BDP.csv', dtype=str)
viviendas = dd.read_csv('/home/guillermo/DATOS/GTVitales y Censo/CENSO GT 2018.csv/VIVIENDA_BDP.csv', dtype=str)
```

```python
quichemunis = [1413,1411,1405,1410,1415, 1401, 1416]
#"Nebaj	Cotzal	Chajul	Cunén	Uspantán"
huehuemunis = [1326, 1318, 1317, 1308, 1315, 1302, 1304, 1307, 1312, 1311]
#"Santa Cruz Barillas	San Mateo Ixtatán	Santa Eulalia	San Pedro Solomá	Todos Santos Cuchumatán	Chiantla	Cuilco	Jacaltenango	La Democracia	La Libertad"
```

```python
personas = fullpersonas[['PCP7', 'PCP12', 'PCP5', 'PCP6', 'MUNICIPIO', 'NUM_VIVIENDA', 'NUM_HOGAR', 'PCP35_A', 'DEPARTAMENTO', 'PCP27']]
```

```python
personas = personas.repartition(npartitions=300)
hogares = hogares.repartition(npartitions=100)
viviendas = viviendas.repartition(npartitions=100)
```

```python
bhogares = hogares[hogares.DEPARTAMENTO.isin(['13','14'])]
bpersonas = personas[personas.DEPARTAMENTO.isin(['13','14'])]
#bviviendas = viviendas[hogares.DEPARTAMENTO.isin(['13','14'])]
```

```python
bpersonas['PCP7'] = bpersonas.PCP7.astype(int)
```

```python
bpersonas.persist()
bhogares.persist()
```

```python
bpersonas['edad0_4'] = (bpersonas.PCP7 >= 0) & (bpersonas.PCP7 <=4)
bpersonas['edad5_9'] = (bpersonas.PCP7 >= 5) & (bpersonas.PCP7 <=9)
bpersonas['edad10_14'] = (bpersonas.PCP7 >= 10) & (bpersonas.PCP7 <=14)
bpersonas['edad15_17'] = (bpersonas.PCP7 >= 15) & (bpersonas.PCP7 <=17)
bpersonas['adultos'] = (bpersonas.PCP7 >= 18)
bpersonas['mujeresERCH'] = (bpersonas.PCP7 >= 15) & (bpersonas.PCP7 <=49) & (bpersonas.PCP6=='2')
bpersonas['hombresERCH'] = (bpersonas.PCP7 >= 15) & (bpersonas.PCP7 <=49) & (bpersonas.PCP6=='1')
```

```python
pre_edades = bpersonas.groupby(['MUNICIPIO', 'NUM_VIVIENDA', 'NUM_HOGAR'])\
    [['edad0_4', 'edad5_9', 'edad10_14', 'edad15_17', 'adultos', 'mujeresERCH', 'hombresERCH']]\
    .sum()
```

```python
def merge_series(series_list):
    output = None
    for series in series_list:
        if output is None:
            output = series.to_frame()
        else:
            output = output.merge(series.to_frame())
    return output
```

```python
dledades = merge_series([
    (pre_edades.edad0_4.astype(float)/pre_edades.mujeresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('e04m'),
    (pre_edades.edad5_9.astype(float)/pre_edades.mujeresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('e59m'),
    (pre_edades.edad10_14.astype(float)/pre_edades.mujeresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('e1014m'),
    (pre_edades.edad15_17.astype(float)/pre_edades.mujeresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('e1517m'),
    (pre_edades.adultos.astype(float)/pre_edades.mujeresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('adultos_m'),
    (pre_edades.edad0_4.astype(float)/pre_edades.hombresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('e04h'),
    (pre_edades.edad5_9.astype(float)/pre_edades.hombresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('e59h'),
    (pre_edades.edad10_14.astype(float)/pre_edades.hombresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('e1014h'),
    (pre_edades.edad15_17.astype(float)/pre_edades.hombresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('e1517h'),
    (pre_edades.adultos.astype(float)/pre_edades.hombresERCH).replace(np.inf, np.nan).dropna().groupby('MUNICIPIO').mean().rename('adultos_h')
])
```

```python
dljefas = bpersonas.assign(jefas=lambda x: (x.PCP5=='1') & (x.PCP6=="2") , mujeres=lambda x: x.PCP6=='2')\
    .groupby(['PCP12', 'MUNICIPIO']).sum()
```

```python
dlhijos = bpersonas[(bpersonas.PCP7 >= 15) & (bpersonas.PCP7 <= 49) & (bpersonas.PCP6=='2') & (bpersonas.PCP35_A != '99') & (bpersonas.PCP35_A != ' ')]\
    .assign(temp=lambda x: x.PCP35_A.astype(int))\
    .groupby(['PCP12', 'MUNICIPIO']).temp.mean()
```

```python
dlhijosB = bpersonas[(bpersonas.PCP7 >= 15) & (bpersonas.PCP7 <= 49) & (bpersonas.PCP6=='2') & (bpersonas.PCP35_A != '99') & (bpersonas.PCP35_A != ' ')]\
    .assign(temp=lambda x: x.PCP35_A.astype(int))\
    .groupby(['MUNICIPIO']).temp.mean()
```

```python
dltenenciav = bpersonas[bpersonas.PCP5=='1'].merge(bhogares, on=['MUNICIPIO', 'NUM_VIVIENDA', 'NUM_HOGAR'], how='inner')\
    .groupby(['PCP12', 'PCH1', 'MUNICIPIO']).count()
```

```python
dlptrabajaA = bpersonas.assign(trabaja=lambda x: x.PCP27=='1', total= lambda x: 1).groupby(['MUNICIPIO', 'NUM_VIVIENDA', 'NUM_HOGAR'])[['trabaja', 'total']].sum()
dlptrabaja = (dlptrabajaA['trabaja']/dlptrabajaA['total']).replace(np.inf, np.nan).dropna().rename('value').reset_index().groupby('MUNICIPIO').value.mean()
```
